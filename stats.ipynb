```r
library(stats)
install.packages("forecast")
library(forecast)
```

```r
data<-read.csv("/content/walmart-sales-dataset-of-45stores.csv")
data$Date <- as.Date(data$Date, format = "%d-%m-%Y")
data$Date
```

```r
my_time_series <- ts(data$Weekly_Sales, start = c(2010,02,05), frequency = 12)
summary(my_time_series)

# Time series plot
plot(my_time_series, main = "Time Series Plot of Weekly Sales")
```

```r
decomposition <- decompose(my_time_series)
summary(decomposition)
plot(decomposition)
```

```r
arima_model <- auto.arima(my_time_series)
forecast_result <- forecast(arima_model, h = 12)  # Forecast for the next 12 periods
summary(arima_model)
plot(forecast_result)
```

```r
accuracy(forecast_result)
```

```r
model <- lm(Weekly_Sales ~ Temperature + CPI + Unemployment + Fuel_Price + Holiday_Flag, data = data)

# Summarize the model
summary(model)
```

```r
data$Predicted_Weekly_Sales <- predict(model)

# Create a scatter plot comparing actual vs. predicted values
ggplot(data, aes(x = Weekly_Sales, y = Predicted_Weekly_Sales)) +
  geom_point(color = 'blue', alpha = 0.5) +  # Actual vs. Predicted scatter plot
  geom_abline(intercept = 0, slope = 1, color = 'red', linetype = 'dashed') +  # Identity line
  labs(title = "Actual vs. Predicted Weekly Sales",
       x = "Actual Weekly Sales",
       y = "Predicted Weekly Sales") +
  theme_minimal()
```

```r
# Load necessary libraries (if not already loaded)
library(caret)

# Set the proportion for the test set (e.g., 20%)
test_proportion <- 0.2

# Determine the index to split the data
split_index <- floor(nrow(data) * (1 - test_proportion))

# Create training and test sets
train_data <- data[1:split_index, ]
test_data <- data[(split_index + 1):nrow(data), ]

# Build a linear regression model on the training data
model <- lm(Weekly_Sales ~ Temperature + CPI + Unemployment + Fuel_Price + Holiday_Flag, data = train_data)

# Summarize the model
summary(model)

# Make predictions on the test data
predictions <- predict(model, newdata = test_data)

# Evaluate model performance on the test data (e.g., calculate RMSE)
install.packages("Metrics")
library(Metrics)
rmse <- rmse(test_data$Weekly_Sales, predictions)
print(paste("Root Mean Squared Error (RMSE):", rmse))
```

```r
plot(test_data$Weekly_Sales, predictions,
     xlab = "Actual Weekly Sales",
     ylab = "Predicted Weekly Sales",
     main = "Actual vs. Predicted Weekly Sales")

# Add a reference line for perfect predictions (y = x)
abline(0, 1, col = "red", lty = 2)
```
